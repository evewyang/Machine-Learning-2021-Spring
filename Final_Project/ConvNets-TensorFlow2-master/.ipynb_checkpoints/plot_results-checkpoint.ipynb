```python
import numpy as np
import pandas as pd
import matplotlib.image as img
import matplotlib.pyplot as plt
```

```python
def plot_accuracy(model_name, model_data):
    train_acc = model_data[:,0]
    test_acc = model_data[:,2]
    # plot epoch vs accuracies
    fig, ax = plt.subplots(figsize=(6, 4))
    dim = np.arange(1,31,1)
    ax.plot(np.append(None,train_acc), color='red', label='train accuracy')
    ax.plot(np.append(None,test_acc), color='green', label= 'test accuracy')
    plt.xticks(dim)
    plt.xlim([0, 30])
    plt.title(model_name)
    plt.xlabel("Epoch")
    plt.ylabel("Accuracy/%")
    ax.legend()
    if (len(train_acc) < 30):
        print("Early stopping at epoch = %d"%(len(train_acc)))
    else:
        print("No early stopping.")
              
    print("Max test accuracy = %s%%"%(np.amax(test_acc)))
    
def plot_loss(model_name, model_data):
    train_loss = model_data[:,1]
    test_loss = model_data[:,3]
    # plot epoch vs accuracies
    fig, ax = plt.subplots(figsize=(6, 4))
    dim = np.arange(1,31,1)
    ax.plot(np.append(None,train_loss), color='blue', label='train loss')
    ax.plot(np.append(None,test_loss), color='orange', label= 'test loss')
    plt.xticks(dim)
    plt.xlim([0, 30])
    plt.title(model_name)
    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    ax.legend()
    print("Min test loss = %s"%(np.amin(test_loss)))
```

```python
GoogLeNet_data = np.loadtxt("results/GoogLeNet.csv", dtype = np.str,delimiter = ",")[1:].astype(np.float32)
VGG11_data = np.loadtxt("results/VGG11.csv", dtype = np.str,delimiter = ",")[1:].astype(np.float32)
VGG13_data = np.loadtxt("results/VGG13.csv", dtype = np.str,delimiter = ",")[1:].astype(np.float32)
VGG19_data = np.loadtxt("results/VGG19.csv", dtype = np.str,delimiter = ",")[1:].astype(np.float32)
ResNet18_data = np.loadtxt("results/ResNet18.csv", dtype = np.str,delimiter = ",")[1:].astype(np.float32)
MobileNet_data = np.loadtxt("results/MobileNet.csv", dtype = np.str,delimiter = ",")[1:].astype(np.float32)
InceptionV3_data = np.loadtxt("results/InceptionV3.csv", dtype = np.str,delimiter = ",")[1:].astype(np.float32)
SEResNet18_data = np.loadtxt("results/SEResNet18.csv", dtype = np.str,delimiter = ",")[1:].astype(np.float32)
SEResNet34_data = np.loadtxt("results/SEResNet34.csv", dtype = np.str,delimiter = ",")[1:].astype(np.float32)
```

```python
plot_accuracy("GoogLeNet",GoogLeNet_data)
plot_loss("GoogLeNet",GoogLeNet_data)
```
